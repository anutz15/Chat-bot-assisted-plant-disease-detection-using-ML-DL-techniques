```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Generate some sample data
np.random.seed(42)
X = np.random.randn(300, 2)
X[:100, :] += 5
X[100:200, :] -= 5
```

```python
# Perform clustering using K-means
kmeans = KMeans(n_clusters=3, random_state=42)
y_pred = kmeans.fit_predict(X)
```

```python
# Plot the data and the clusters
plt.scatter(X[y_pred == 0, 0], X[y_pred == 0, 1], color='red', label='Cluster 1')
plt.scatter(X[y_pred == 1, 0], X[y_pred == 1, 1], color='green', label='Cluster 2')
plt.scatter(X[y_pred == 2, 0], X[y_pred == 2, 1], color='blue', label='Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='black', label='Centroids')
plt.title('K-means clustering with 3 clusters')
plt.legend()
plt.show()
```
